```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
```

```python
g=9.81 #m/s2
A1, A2, A3 = 1e-2, 1e-2, 1e-2
Ac1, Ac2, Ac3 = 1.5, 2, 2.5

Qin0 = 0.03 #m3/s

h10 = ((Qin0/A1)**2)/(2*g)
h20 = ((Qin0/A2)**2)/(2*g)
h30 = ((Qin0/A3)**2)/(2*g)
h3sp = 1.1*h30
kc=0.25

def rhs(t, v):
    h1, h2, h3 = v
    err = h3sp - h3
    Qin=Qin0 + kc*err

    q1 = A1*np.sqrt(2*g*h1)
    q2 = A2*np.sqrt(2*g*h2)
    q3 = A3*np.sqrt(2*g*h3)
    return np.array([ (Qin - q1)/Ac1,
                     (q1-q2)/Ac2,
                      (q2-q3)/Ac3])
```

```python
tend = 5000.
res=solve_ivp(rhs, (0,tend), np.array([h10, h20, h30]), dense_output=True, method='Radau')
```

```python
tplot=np.linspace(0.,tend, 1000)
h1,h2,h3=res.sol(tplot)
```

```python
fig=make_subplots()
fig.add_scatter(x=tplot, y=h1, name='tank1')
fig.add_scatter(x=tplot, y=h2, name='tank2')
fig.add_scatter(x=tplot, y=h3, name='tank3')
fig.update_layout(width=800, height=600, template='plotly_dark')
```
